```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
data =pd.read_csv(r"/kaggle/input/credit-card-customer-churn-prediction/Churn_Modelling.csv")
data.head(3)
```

```python
data.info()
```

```python
#to check duplicated rows 
data.duplicated().sum()
```

```python
#to check number of people who exited the bank and not exited 
data['Exited'].value_counts()
```

```python
data.drop(columns=['RowNumber','CustomerId','Surname'],inplace=True)
```

```python
data
```

```python
# onhot=OneHotEncoder()
# onhot.fit_transform(data,columns=['Geography','Gender'],drop_first=True)
#df=pd.get_dummies(data,columns=['Geography','Gender'],drop_first=True)


# from sklearn.preprocessing import OneHotEncoder
# from sklearn.compose import ColumnTransformer

# # Example: Assume 'data' is a DataFrame
# categorical_features = ['Geography', 'Gender']

# column_transformer = ColumnTransformer(
#     transformers=[
#         ('encoder', OneHotEncoder(drop='first'), categorical_features)
#     ],
#     remainder='passthrough'  # keeps the other columns
# )

# transformed_data = column_transformer.fit_transform(data)



# Create dummy variables
data = pd.get_dummies(data, columns=['Geography', 'Gender'], drop_first=True)

# Convert boolean columns to integers (0 and 1)

```

```python
data=data.astype(int)
```

```python
data
```

```python
x=data.drop(columns=['Exited'])
y=data['Exited']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

# y_train_scaled=scaler.fit_transform(y_train)
# y_test_scaled=scaler.transform(y_test)
# never try to scale the target data as here y is target data , 
# if you scale it affects the result as well
```

```python
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python

import tensorflow
from tensorflow import keras 
from tensorflow.keras import Sequential 
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.regularizers import l2
```

```python
model=Sequential()

model.add(Dense(3,activation='sigmoid',input_dim=11))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

Modal compilation 

```python
model.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
model.fit(x_train_scaled,y_train,epochs=10)
```

```python
model.layers[0].get_weights()
```

```python
log_pred=model.predict(x_test_scaled)
```

```python
log_pred
```

```python
y_pred=np.where(log_pred>0.5,1,0)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

changes to enhance model

```python
model=Sequential()

model.add(Dense(10,activation='relu',input_dim=11,kernel_regularizer=l2(0.001)))
         
model.add(Dense(16,activation='relu'))

model.add(Dense(10,activation='relu',kernel_regularizer=l2(0.001)))
model.add(Dropout(0.5))
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer=keras.optimizers.Adam(learning_rate=0.0008),metrics=['accuracy'])

history=model.fit(x_train_scaled,y_train,epochs=20,validation_split=0.2)
```

```python
model.layers[0].get_weights()
```

```python
log_pred=model.predict(x_test_scaled)
```

```python
y_pred=np.where(log_pred>0.5,1,0)
y_pred
```

```python
accuracy_score(y_test,y_pred)
```
